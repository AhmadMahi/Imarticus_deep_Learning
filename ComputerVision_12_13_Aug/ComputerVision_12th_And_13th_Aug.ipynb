```python
rose = cv2.imread("blue_roses.jpg")
cv2_imshow(rose)
```

```python
import matplotlib.pyplot as plt
```

```python
rose_hsv = cv2.cvtColor(rose, cv2.COLOR_BGR2HSV)
cv2_imshow(rose_hsv)

lower_blue = (60,30, 30)
upper_blue= (150, 255, 255)

mask = cv2.inRange(rose_hsv, lower_blue, upper_blue)
plt.figure()
plt.imshow(mask, cmap= plt.cm.binary)

masked_rose  = cv2.bitwise_and(rose, rose, mask = mask)
plt.figure()
plt.imshow(masked_rose[:,:,::-1])
```

```python
import numpy as np
red_rose = cv2.imread("red_rose.jpg")[:,:,::-1]
translated_rose = cv2.warpAffine(red_rose, np.array([[1, 0, 50], [0., 1, -50]]), (red_rose.shape[1], red_rose.shape[0]))
plt.imshow(translated_rose)
```

```python
H, W = red_rose.shape[:-1]
rot_center = (W/2, H/2)
rot_mat = cv2.getRotationMatrix2D(center = rot_center , angle= 90, scale= 0.5)
rotated_rose = cv2.warpAffine(red_rose, rot_mat, (W, H))
plt.imshow(rotated_rose)
```

```python

sudoku = cv2.imread("test.jpg")[:,:,::-1]
width = 610
height = 405

sudoku = cv2.resize(sudoku,(width,height))

original_xys = np.float32([[183, 255], [292, 199], [236, 301], [355, 238]])
mapped_xys = np.float32([[0, 0], [500, 0], [0, 500], [520, 520]])

M = cv2.getPerspectiveTransform(original_xys, mapped_xys)
perspective_trans = cv2.warpPerspective(sudoku, M, (520, 520))

M = cv2.getAffineTransform(original_xys[:-1, :], mapped_xys[:-1, :])
affine_trans = cv2.warpAffine(sudoku, M, (520, 520))

fig, axes = plt.subplots(1, 3, figsize=(3 * 4, 4))
axes[0].imshow(sudoku)
axes[0].autoscale(False)
axes[0].set_title("original")
axes[0].plot(original_xys[:, 0], original_xys[:, 1], "rs")

axes[1].imshow(perspective_trans)
axes[1].autoscale(False)
axes[1].plot(mapped_xys[:, 0], mapped_xys[:, 1], "rs")
axes[1].set_title("perspective transformation")

axes[2].imshow(affine_trans)
axes[2].autoscale(False)
axes[2].plot(mapped_xys[:, 0], mapped_xys[:, 1], "rs")
axes[2].set_title("affine transformation")

plt.show()
```

```python

image = cv2.imread("Square.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
corners = cv2.goodFeaturesToTrack(gray, maxCorners=4, qualityLevel=0.01, minDistance=10)
corners = np.int0(corners)

for corner in corners:
    x, y = corner.ravel()
    cv2.circle(image, (x, y), 5, (0, 255, 0), -1)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
